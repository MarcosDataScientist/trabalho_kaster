# Carregamento de Dados: Feriados Nacionais (2020 - 2026)

Este fluxo realiza consultas à API `brasilapi.com.br` para coletar os feriados nacionais do Brasil entre os anos de 2020 e 2026.
Os dados consolidados são salvos em formato JSON na pasta `Dados`.

```python
import requests
import json
import os
import time

anos = range(2020, 2027)
todos_feriados = []

for ano in anos:
    url = f"https://brasilapi.com.br/api/feriados/v1/{ano}"
    response = requests.get(url)
    
    if response.status_code == 200:
        feriados_ano = response.json()
        todos_feriados.extend(feriados_ano)
        print(f"Feriados de {ano} carregados com sucesso.")
    else:
        print(f"Erro ao buscar feriados de {ano}. Status code: {response.status_code}")
    
    # Pausa para nao sobrecarregar a API
    time.sleep(0.5)

diretorio_dados = "Dados_brutos"
if not os.path.exists(diretorio_dados):
    os.makedirs(diretorio_dados)

caminho_arquivo = os.path.join(diretorio_dados, "feriados_nacionais.json")

with open(caminho_arquivo, "w", encoding="utf-8") as f:
    json.dump(todos_feriados, f, ensure_ascii=False, indent=4)

print(f"Dados salvos com sucesso em {caminho_arquivo}")
```
