```python

# Carregar pacotes
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
#cirando caminho (nao usamos mas pode ser bom pra outras maquinas)
caminho = os.getcwd()
#criacao dataframe remove cabecalho e as duas primeiras linhas sem uso
df_acidentes_totais = pd.read_excel("acidentes_transito_total.xlsx", header=None,skiprows=2)
for col in df_acidentes_totais.columns:
    print(repr(col))

colunas_texto = [
    "Município/Estado",
    "Região a que Pertence"
]

colunas_numericas = [
    col for col in df_acidentes_totais.columns
    if col not in colunas_texto
]

df_acidentes_totais[colunas_numericas] = (
    df_acidentes_totais[colunas_numericas]
    .replace("-", 0)
    .replace(r"\.", "", regex=True)
    .apply(pd.to_numeric, errors="coerce")
    .fillna(0)
    .astype(int)
)

display(df_acidentes_totais)

df_acidentes_totais.info()
```
